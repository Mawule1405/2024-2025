```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import warnings

# Ignorer l'avertissement spécifique
warnings.filterwarnings("ignore", message="X does not have valid feature names, but KMeans was fitted with feature names")
```

# 1. Kmeans avec des données Saisies

```python
# Exemple de données d'entrée non étiquetées
X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]])
# Création du modèle de clustering K-means
kmeans = KMeans(n_clusters=2, n_init='auto',random_state=123)
# Entraînement du modèle sur les données
kmeans.fit(X)
#Précsion du modèle avec le score ou l'inertie
print(kmeans.score(X))
print(kmeans.inertia_)
# Prédiction des clusters pour les données d'entrée
y_pred = kmeans.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1], c=y_pred)
plt.show()
```

```python
centres=kmeans.cluster_centers_
centres
```

```python
plt.scatter(X[:,0],X[:,1], c=y_pred)
plt.scatter(centres[:,0], centres[:,1], c="r")
plt.show()
```

```python
Lpred=np.array([[6,9],[5,6],[4,3],[6,6],[2,2]])
pred=kmeans.predict(Lpred)
pred
```

```python
plt.scatter(X[:,0],X[:,1], c=y_pred)
plt.scatter(centres[:,0], centres[:,1], c="r")
plt.scatter(Lpred[:,0],Lpred[:,1], c=pred)
plt.show()
```

# 2. Kmeans avec des données importées

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings

# Ignorer l'avertissement spécifique
warnings.filterwarnings("ignore", message="X does not have valid feature names, but KMeans was fitted with feature names")
```

```python
df=pd.read_excel("données/kvoisins.xlsx",sheet_name="parti")
df
```

```python
X=df.values
X
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto', random_state=123)
# Entraînement du modèle sur les données
kmeans.fit(df)
#Précsion du modèle avec le score ou l'inertie
print(kmeans.score(df))
print(kmeans.inertia_)
# Prédiction des clusters pour les données d'entrée
y_pred = kmeans.predict(df)
```

```python
plt.scatter(df.X,df.Y, c=y_pred)
plt.show()
```

```python
centres=kmeans.cluster_centers_
centres
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df.X,df.Y, c=y_pred)
plt.scatter(centres[:,0], centres[:,1], c="r")
plt.show()
```

```python
Lpred=np.array([[6,9],[5,6],[4,3],[6,6],[2,2]])
pred=kmeans.predict(Lpred)
pred
```

```python
plt.scatter(X[:,0],X[:,1], c=y_pred)
plt.scatter(centres[:,0], centres[:,1], c="r")
plt.scatter(Lpred[:,0],Lpred[:,1], c=pred)
plt.show()
```

# K means avec Iris

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
X, y = load_iris(return_X_y=True)
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
# Entraînement du modèle sur les données
kmeans.fit(X)
print(kmeans.inertia_)
# Prédiction des clusters pour les données d'entrée
y_pred = kmeans.predict(X)
centres=kmeans.cluster_centers_
```

```python
plt.scatter(X[:,0],X[:,1], c=y_pred)
plt.show()
```

```python
plt.scatter(X[:,0],X[:,1], c=y_pred)
plt.scatter(centres[:,0], centres[:,1], c="r")
```

```python
kmeans.predict([[6.5, 3.2 , 1.6, .5]])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y,y_pred )
```

# Courbe à Coude

```python
K=range(1,6)
L=[]
for k in K:
    kmeans = KMeans(n_clusters=k, n_init="auto")
# Entraînement du modèle sur les données
    kmeans.fit(X)
    L.append(kmeans.inertia_)

plt.figure(figsize=(10,8))
plt.plot(K, L)
plt.xticks(K)
plt.show()
```
