```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Charger le jeu de données Iris
data = load_iris()
X = data.data
features = data.feature_names

# Ajuster le modèle PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Calculer les composantes principales
pcs = pca.components_

# Créer le cercle des corrélations
plt.figure(figsize=(8, 8))

# Tracer le cercle
circle = plt.Circle((0, 0), 1, color='gray', fill=False)
plt.gca().add_artist(circle)

# Tracer les vecteurs
for i, (x, y) in enumerate(pcs.T):
    plt.arrow(0, 0, x, y, head_width=0.05, head_length=0.05, fc='red', ec='red')
    plt.text(x, y, features[i], color='black', ha='center', va='center')

# Limites des axes
plt.xlim(-1.1, 1.1)
plt.ylim(-1.1, 1.1)

# Étiquettes et titre
plt.xlabel('Composante Principale 1')
plt.ylabel('Composante Principale 2')
plt.title('Cercle des corrélations')
plt.grid()
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Générer des données d'exemple
np.random.seed(0)
data = np.random.rand(10, 10)

# Liste des palettes similaires à 'viridis'
palettes = ['viridis', 'plasma', 'inferno', 'magma', 'cividis', 'turbo']

# Créer des sous-graphiques pour chaque palette
fig, axes = plt.subplots(1, len(palettes), figsize=(20, 3))

for ax, palette in zip(axes, palettes):
    cax = ax.matshow(data, cmap=palette)
    ax.set_title(palette)
    fig.colorbar(cax, ax=ax)

plt.show()
```
